```python
import pandas as pd
import pandas_datareader.data as web
from pandas import Series, DataFrame
import datetime
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
start = datetime.datetime(1990, 1, 1)
end = datetime.datetime(2021, 5, 5)
df = web.DataReader("MSFT", 'yahoo', start, end)
```

```python
close_px = df['Close']
MA5 = close_px.rolling(window = 5).mean()
MA20 = close_px.rolling(window = 20).mean()
MA100 = close_px.rolling(window = 100).mean()
MA200 = close_px.rolling(window = 200).mean()
```

```python
mpl.rc('figure', figsize=(15, 8))
mpl.__version__
style.use('ggplot')
close_px.plot(label = 'MSFT')
MA5.plot(label = 'MA5')
MA20.plot(label = 'MA20')
MA100.plot(label = 'MA100')
MA200.plot(label = 'MA200')
plt.ylabel('Close Price in USD')
plt.suptitle('Close price VS Moving Average 5/20/100/200 days of MSFT')
plt.legend()
```

```python
#Preprocess data
import numpy as np
#Set index as datetime
df.index = pd.to_datetime(df.index)
#Set label by shifting Closing price by -1
df['Label'] = df['Close'].shift(-1)
#Log_r
log_r = np.log(df['Close'] / df['Close'].shift(1))
max_roll_window = 0

#Drop non-used columns
df = df.drop(['Open', 'High', 'Low', 'Volume', 'Adj Close'], axis = 1)

#Drop NaN values
df = df.dropna()

#Standarizing
df = (df - df.mean())/df.std()
    
    
## rolling mean
close_px = df['Close']
#mavg = close_px.rolling(window=100).mean()
df['MA5'] = close_px.rolling(5,min_periods=1).mean()
df['MA20'] = close_px.rolling(20,min_periods=1).mean()
df['MA100'] = close_px.rolling(100,min_periods=1).mean()
df['MA200'] = close_px.rolling(200,min_periods=1).mean()
print(df.tail())

%matplotlib inline
import matplotlib.pyplot as plt


# MA5 and close price in 2020 

df['Close'].loc['2020-01-01':'2020-12-29'].plot(figsize=(12,8), label = 'Close')
df['MA5'].loc['2020-01-01':'2020-12-29'].plot(figsize=(12,8),label = 'MA5')
plt.ylabel('Close Price in USD')
plt.suptitle('Moving Average 5 days of MSFT in 2020')
plt.legend()
```
